* Data: Soil samples from various geographical regions in Africa and related MIR, spatial and geographical properties.
* Unknown: Values of Ca, P, Sand, SOC, pH on unseen samples.
* Condition: MCRMSE ( Mean Column Root Mean Squared Error ) Minimize the evaluation metric.

** Questions **

1. Is the data enough to accurately predict the values ?
2. Is the training and test samples are from the same distribution, or test samples have properties that we haven't seen in the training set ?
3. Is MCRMSE the right evaluation metric to minimize ?
4. Do we need some auxiliary data that will answer this question well ?
5. How to set up the cross validation scheme that accurately captures the real training and test sample scenario ?

* Data have been monotonously transformed from the original measurements and thus include negative values.
* Spatial predictors have been mean centered and scaled.

```python
def pca(X, n_components=2):
    """
    Principle Component Analysis
    """
    
    pca = PCA(n_components=n_components, whiten=True)
    return pca.fit_transform(X)


def t_sne(X, n_components=2):
    """
    t-stochastuc neighbor embedding
    """
    
    tsne = TSNE(n_components=n_components, random_state=2, init='pca')
    return tsne.fit_transform(X)
```

** Utility that would transform a wavelength band down to number of components mentioned. **

```python
bands = {
    '7000': (1, 260),
    '6000': (260, 778),
    '5000': (778, 1297),
    '4000': (1297, 1815),
    '3000': (1815, 2334),
    '2000': (2334, 2852),
    '1000': (2852, 3371),
    '<1000': (3371, 3579)
    
}

def transform(df, band, columns):
    
    start, end = bands[band]
    df_sub  = df[columns[start:end]]
    
    n_components = 2
    tsne = TSNE(n_components=n_components, init='pca', random_state=10)
    
    df_sub_tsne = tsne.fit_transform(df_sub)
    
    column_names = [band + '_' + str(i) for i in range(n_components)]
    df_sub_tsne = pd.DataFrame(df_sub_tsne, columns=column_names)
    
    return df_sub_tsne

def concat(df_list):
    return pd.concat(df_list, axis=1) # along columns

def transform_and_concat(df, columns):
    transformed = [transform(df, k, columns) for k in bands.keys()]
    return concat(transformed)
```

** Objective of this function. **

Way of calculating relative strength of each feature by first binning them in equal proportion and then calculating the $R^2$ with the target variable

* Function that would bin the variable.
* Function that would calculate the $R^2$ with a given target variable.

** This strategy was discussed in one of the winning solutions. **

```python
def calculate_r_squared(X, y):
    slope, intercept, r_value, p_value, std_err = linregress(X, y)
    return r_value ** 2

def calculate_score(X, y, bins):
    X_agg = []
    y_agg = []
    
    for bin_ in range(bins.nunique()):
        mask = (bins == bin_)

        X_agg.append(X.loc[mask].mean())
        y_agg.append(y.loc[mask].mean())
    
    X_agg = np.array(X_agg)
    y_agg = np.array(y_agg)
    
    r_squared = calculate_r_squared(X_agg, y_agg)
    
    return r_squared
    
    
def feature_r_squared(X, y, features, bins):
    return [calculate_score(X[feature], y, bins) for feature in features]
    
def feature_relevance(X, y, nbins=8):
    
    features = X.columns
    bins = pd.cut(X[features[0]], nbins, labels=False)
     
    return feature_r_squared(X, y, features, bins)
```

```python
feature_relevance_Ca = feature_relevance(train_, train.Ca)
feature_relevance_P = feature_relevance(train_, train.P)
feature_relevance_Sand = feature_relevance(train_, train.Sand)
feature_relevance_SOC = feature_relevance(train_, train.SOC)
feature_relevance_pH = feature_relevance(train_, train.pH)
```
